```python
# Read data from the file and handle null values

import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
df.columns
```

```python
# Rename columns names .. make them lower case and replace space with underscore

df.rename(columns={'Order Id':'order_id', 'City':'city'})

df.columns = df.columns.str.lower()

df.columns = df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

```python
# Derive 3 new columns discount, sale price and profit

df['discount'] = df['list_price'] * df['discount_percent'] * .01

df['sale_price'] = df['list_price'] - df['discount']

df['profit'] = df['sale_price'] - df['cost_price']

df
```

```python
df.info()
```

```python
# Convert order date from object data type to datetime

df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# Drop columns cost price, list price and discount percent

df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
# Load the data into sql server

import sqlalchemy as sal
engine = sal.create_engine('mssql://MAERON-7K/EDA?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect()
```

```python
# Load the data into sql server using replace option if table with same name exists

df.to_sql('df_orders', con=conn , index=False, if_exists = 'replace')
```
